```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
img = cv2.imread('Img_01.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
# Step 2: Create a blurred (unsharp) version of the original image
unsharp = cv2.GaussianBlur(img, (9, 9), 10.0)

# Step 3: Create the difference image (original − unsharp)
diff = cv2.subtract(img.astype(np.float32), unsharp.astype(np.float32))

# Step 4: Apply USM to get the resulting image using `sharpened = original + (original − unsharp) × amount`
amount = 1.5  # Reasonable starting value for amount
sharpened = cv2.add(img.astype(np.float32), diff * amount)

# Ensure pixel values are within the valid range [0, 255]
sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)

# Display the sharpened image
plt.imshow(sharpened)
```
